Dataset on the job outcomes of students who graduated from college between 2010 and 2012. 

Following are the attributes 
1. Rank - Rank by median earnings (the dataset is ordered by this column).
2. Major_code - Major code.
3. Major - Major description.
4. Major_category - Category of major.
5. Total - Total number of people with major.
6. Sample_size - Sample size (unweighted) of full-time.
7. Men - Male graduates.
8. Women - Female graduates.
9. ShareWomen - Women as share of total.
10. Employed - Number employed.
11. Median - Median salary of full-time, year-round workers.
12. Low_wage_jobs - Number in low-wage service jobs.
13. Full_time - Number employed 35 hours or more.
14. Part_time - Number employed less than 35 hours.



```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
recent_grads = pd.read_csv("recent-grads.csv", encoding = "UTF-8")
recent_grads.iloc[0]
recent_grads.head()
recent_grads.describe()
recent_grads = recent_grads.dropna(0)
recent_grads.info()
```

**Do students that majored in subjects that were majority female make more money?**

It is evident from the plot that,wherever sharewoman>0.5(which would mean, woman are the majority). The median pay is not at his highest. Infact the highest values come in between the 0.15 to 0.4 range

Observe the below graph to inform us of the number of men and women in each catagory of majors.

```python
women_list= {}
men_list = {}
x = recent_grads["Major_category"].unique()
for val in x:
    grouped_cat = recent_grads[recent_grads["Major_category"]== val]
    men = grouped_cat["Men"].sum()
    women = grouped_cat["Women"].sum()
    men_list[val] = men
    women_list[val] = women
print(men_list)
print("\n")
print(women_list)

new_df = pd.DataFrame(list(men_list.items()),columns = ['Major_category','Men'])
new_df["Women"] = women_list.values()
new_df
ax = new_df[['Men', 'Women']].plot.bar()
ax.set_xticklabels(new_df["Major_category"])
```

The median salary is abour 35-37k USD and it s not evenly distributed as the red line. a quarter of the salaries lie in the 2nd quartile between 32-36k 

The unemployemnt rate seems to be evenly distributed
